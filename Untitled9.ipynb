```python
crm= pd.read_excel('samplekkm.xlsx')
#drop data with100% missing values
crm.drop(['LocNames','Unnamed: 18','Unnamed: 19','AllegationUnfounded'], axis=1, inplace=True)
crm.head()
```

```python
#split evnttype into cause and reason
crm2 = crm.copy()
crm2['Cause'] = crm2.EventType.str.split('(').str.get(0)
crm2['Reason'] = crm2.EventType.str.split('(').str.get(1) 
crm2.drop(['EventType'], axis=1, inplace=True)

crm2['Cause'] = crm2['Cause'].str.replace('/', ' or ')

#fill missing values 
df1 = crm2[['OccuranceTime']]
df1.fillna("00:00:00",inplace=True)
crm2['OccuranceTime']=df1
df2=crm2[['Cause']]
df2.fillna("Unknown",inplace=True)
crm2['Cause']=df2
df3=crm2[['zone']]
df3.fillna("0",inplace=True)
crm2['zone']=df3
df3=crm2[['LocCode']]
df3.fillna("0",inplace=True)
crm2['zone']=df3
#Merge dataframes
crm2['OccuranceTime']=crm2['OccuranceTime'].astype('str').str[:2]
crm2['OccuranceTime']=crm2['OccuranceTime'].astype('category')
#split date into weekdays and weekends
crm2['dayNum'] = crm2['OccuranceDate'].dt.dayofweek
crm2['isWeekday']=crm2['dayNum'].empty
crm2.loc[crm2.dayNum<5, 'isWeekday'] = True
crm2.loc[crm2.dayNum>4, 'isWeekday'] = False


#Assign contributr
crm2['LocType']=crm2['LocType'].astype('category')
crm2['LocCode']=crm2['LocCode'].astype('category')
crm2['IncidentReporterTitle']=crm2['IncidentReporterTitle'].astype('category')
crm2['zone'] = crm2.zone.astype('category')
crm2['occuredat']=crm2.occuredat.astype('category')
crm2['Cause']=crm2.Cause.astype('category')

#drop columns from crm2
crm2.drop(['OtherIssues','EventPlace','numofnotes','DateOfMostRecentNote','numofaction','DateOfMostRecentAction','numofemails','DateOfMostRecentEmail'], axis=1, inplace=True)
crm2.drop(['Reason','LocType','ESC','IncidentStatus','OccuranceDate'], axis=1, inplace=True)
crm2.drop(['CreateDate'], axis=1, inplace=True)
targetVar='Cause'
crm2[targetVar]=crm2[targetVar].astype('category')

print(crm2.info())
```

```python
from sklearn.model_selection  import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.ensemble import BaggingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
print('Done')
```

```python
RANDOM_STATE  = 3812 #random state to produce consistent results. 3812 is randomly chosen number here
targetVar='Cause'
crm2[targetVar]=crm2[targetVar].astype('category')
myData = crm2.copy()
features= myData[myData.columns.difference([targetVar])]
features = pd.get_dummies(features)
target = myData[targetVar]
X_new = SelectKBest(chi2, k=150).fit_transform(features, target)
X_train, X_test, y_train, y_test = train_test_split(features, target,test_size=0.20,random_state =RANDOM_STATE)
X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train,test_size=0.125,random_state =RANDOM_STATE)

print('Done')
```

```python
def getAccuracy(classifier):
    boostedClassifier = make_pipeline(StandardScaler(),classifier)
    boostedClassifier.fit(X_train, y_train)
    pred_Dev_Data = boostedClassifier.predict(X_dev)
    devAccScore = metrics.accuracy_score(y_dev, pred_Dev_Data)
    return devAccScore
print('done')
```

```python
depthArr = [1,2,5,10,15]
bestRFdepth=0
bestAcc=0
for i in range(len(depthArr)):
    randomForest_clf = RandomForestClassifier(n_jobs=100,random_state=RANDOM_STATE, max_depth=depthArr[i])
    devAccScore=getAccuracy(randomForest_clf)
    print(devAccScore)
    if devAccScore>bestAcc:
        bestAcc=devAccScore
        bestRFdepth=depthArr[i]
print("Best depth for random forest is: ",end="");print(bestRFdepth)
print("accScore is: " + str(bestAcc))

depthArr = [1,2,5,10,15]
bestDTdepth=0
bestAcc=0
for i in range(len(depthArr)):
    decisionTree_clf = DecisionTreeClassifier(max_depth=depthArr[i],random_state=RANDOM_STATE)
    devAccScore=getAccuracy(decisionTree_clf)
    print(devAccScore)
    if devAccScore>bestAcc:
        bestAcc=devAccScore
        bestDTdepth=depthArr[i]
print("Best depth for DT is: ",end="");print(bestDTdepth)
print("accScore is: " + str(bestAcc))
```

```python
gammaArr = [0.001,0.01,0.1,0.2]
bestSVMGamma=0
bestAcc=0
for i in range(len(gammaArr)):
    svc_clf=SVC(gamma=gammaArr[i],random_state=RANDOM_STATE)
    devAccScore=getAccuracy(svc_clf)
    print(devAccScore)
    if devAccScore>bestAcc:
        bestAcc=devAccScore
        bestSVMGamma=gammaArr[i]
print("Best gamma for SVM is: ",end="");print(bestSVMGamma)
print("accScore is: " + str(bestAcc))

```
